```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Câu 1:Tạo DataFrame df chứa danh sách các sinh viên khoa CN, khoa TO và khoa VL. Các danh sách 
#này được đọc từ các tập tin sinh_vien_CN.csv, sinh_vien_TO.csv và sinh_vien_VL.csv

CN=pd.read_csv("sinh_vien_CN.csv")
TO=pd.read_csv("sinh_vien_TO.csv")
VL=pd.read_csv("sinh_vien_VL.csv")
df=pd.DataFrame(TO)
df_1=pd.DataFrame(CN)
df_2=pd.DataFrame(VL)
df=df.append(df_1)
df=df.append(df_2)
df
```

```python
#Câu 2:Cho biết số cột, số dòng, các dữ liệu head và tail của DataFrame df.
#Số cột
print("Số cột :",df.shape[1])
#Số hàng
print("Số hàng :",df.shape[0])
# dữ liệu head
df.head()
```

```python
#dữ liệu tail
df.tail()
```

```python
#Câu 3: Tạo lại index cho DataFrame df với giá trị bắt đầu từ 0
df.reset_index(inplace=True)
df.pop("index")
df
```

```python
# Câu 4 : Tạo DataFrame df_ket_qua từ tập tin ket_qua.csv
KQ=pd.read_csv("ket_qua.csv")
df_ket_qua=pd.DataFrame(KQ)
df_ket_qua.head(5)
```

```python
# Câu 5 : Tạo DataFrame df_dtb bằng cách nhóm df_ket_qua theo masv và tính điểm trung bình cho mỗi sinh viên
df_dtb=df_ket_qua.groupby("masv").mean()
df_dtb.reset_index(inplace=True)
df_dtb.head(5)
```

```python
# Bài 6 : Trộn df và df_dtb theo masv và có kết quả của df như sau:
df_merge=pd.merge(df_dtb,df)
df_merge=df_merge[["masv","ho","ten","gioitinh","ngaysinh","email","didong","cmnd","hocbong","makh","diem"]]
df_merge.head(5)
```

```python
# Bài 7 : In ra các sinh viên của khoa CN và có diem>8
df_khoaCN=df_merge[(df_merge["makh"]=="CN") & (df_merge["diem"]>8)]
df_khoaCN.head(2)
```

```python
# Bài 8 : In ra các sinh viên của khoa CN và có diem>8. Sắp xếp tăng theo ten và giảm theo diem
df_khoaCN_=df_merge[(df_merge["makh"]=="CN") & (df_merge["diem"]>8)]
df_khoaCN_=df_khoaCN_.sort_values(["ten","diem"],ascending=[True, False])
df_khoaCN_.head(5)
```

```python
# Bài 9 : In ra các nam sinh viên không có học bổng.
df_hb=df_merge[(df_merge["gioitinh"]==True) & (df_merge["hocbong"]==0)]
df_hb.head(3)
```

```python
#Bài 10 :In ra tổng học bổng của mỗi khoa
df_hbk=df_merge[{"makh","hocbong"}]
df_hbk_1=df_hbk.groupby("makh").sum()
df_hbk_1.reset_index(inplace=True)
df_hbk_1
```

```python
# # Bài 11 : Thống kê học bổng của mỗi khoa.
df_hbk=df_merge[{"makh","hocbong"}]

df_hbk_1=df_hbk.groupby("makh").sum()
df_hbk_2=df_hbk.groupby("makh").max()
df_hbk_3=df_hbk.groupby("makh").min()
df_hbk_4=df_hbk.groupby("makh").mean()

df_hbk_1.rename(columns={'hocbong':'sum'}, inplace=True)
df_hbk_1.reset_index(inplace=True)
df_hbk_2.rename(columns={'hocbong':'max'}, inplace=True)
df_hbk_2.reset_index(inplace=True)
df_hbk_3.rename(columns={'hocbong':'min'}, inplace=True)
df_hbk_3.reset_index(inplace=True)
df_hbk_4.rename(columns={'hocbong':'mean'}, inplace=True)
df_hbk_4.reset_index(inplace=True)

df_hbk_=pd.merge(df_hbk_1,df_hbk_2)
df_hbk_=pd.merge(df_hbk_,df_hbk_3)
df_hbk_=pd.merge(df_hbk_,df_hbk_4)

df_hbk_
```

```python
# Bài 12 : Thống kê điểm của mỗi khoa.
df_hbk=df_merge[{"makh","diem"}]

df_hbk_1=df_hbk.groupby("makh").count()
df_hbk_2=df_hbk.groupby("makh").max()
df_hbk_3=df_hbk.groupby("makh").min()
df_hbk_4=df_hbk.groupby("makh").mean()

df_hbk_1.rename(columns={'diem':'count'}, inplace=True)
df_hbk_1.reset_index(inplace=True)
df_hbk_2.rename(columns={'diem':'max'}, inplace=True)
df_hbk_2.reset_index(inplace=True)
df_hbk_3.rename(columns={'diem':'min'}, inplace=True)
df_hbk_3.reset_index(inplace=True)
df_hbk_4.rename(columns={'diem':'mean'}, inplace=True)
df_hbk_4.reset_index(inplace=True)

df_hbk_=pd.merge(df_hbk_1,df_hbk_2)
df_hbk_=pd.merge(df_hbk_,df_hbk_3)
df_hbk_=pd.merge(df_hbk_,df_hbk_4)

df_hbk_
```

```python
# Bài 13 : Cho biết các sinh viên có điểm lớn nhất, điểm nhỏ nhất.
df_merge[df_merge["diem"]==df_merge["diem"].min()]
```

```python
df_merge[df_merge["diem"]==df_merge["diem"].max()]
```

```python
# Bài 14 : Cho biết 5 sinh viên có điểm lớn nhất của mỗi khoa, sắp tăng theo mã khoa và giảm theo điểm
df_khoaCN=df_merge[(df_merge["makh"]=="CN")]
df_khoaCN=df_khoaCN.sort_values(by="diem",ascending=False)
df_khoaCN=df_khoaCN.head(5)

df_khoaTO=df_merge[(df_merge["makh"]=="TO")]
df_khoaTO=df_khoaTO.sort_values(by="diem",ascending=False)
df_khoaTO=df_khoaTO.head(5)

df_khoaVL=df_merge[(df_merge["makh"]=="VL")]
df_khoaVL=df_khoaVL.sort_values(by="diem",ascending=False)
df_khoaVL=df_khoaVL.head(5)

df_khoa=df_khoaCN.append(df_khoaTO)
df_khoa=df_khoa.append(df_khoaVL)

df_khoa=df_khoa.sort_values(["makh","diem"],ascending=[True, False])
df_khoa
```

```python
# Bài 15 : Trong DataFrame df tạo thêm cột ketqua với kết quả là giỏi nếu điểm>=9, kết quả là khá nếu
# điểm<9 và điểm >=6.5, kết quả là trung bình nếu điểm <6.5 và điểm >=5, kết quả là yếu nếu điểm<5.
kq=np.array(["Giỏi",'Khá',"Trung bình","Yếu"])
def so_sanh(x):
    if(x>=9):
        return kq[0]
    elif(x>=6.5):
        return kq[1]
    elif(x >= 5):
        return kq[2]
    else :
        return kq[3]

df_temp=df_merge.copy()
df_temp.drop(df_temp.columns[[3,4,5,6,7,8]], axis=1, inplace=True)
df_temp["Ketqua"]=df_temp["makh"]
df_temp["Ketqua"]=df_temp.apply(lambda x: so_sanh(x["diem"]),axis=1)
df_temp.head(7)
```

```python
# Bài 16 : Trong mỗi khoa, đếm số sinh viên theo các loại kết quả
df_CN = df_temp[df_temp["makh"]=="CN"]
df_TO = df_temp[df_temp["makh"]=="TO"]
df_VL = df_temp[df_temp["makh"]=="VL"]

Gioi = [(df_CN["Ketqua"]=="Giỏi").sum(),(df_TO["Ketqua"]=="Giỏi").sum(),(df_VL["Ketqua"]=="Giỏi").sum()]
Kha = [(df_CN["Ketqua"]=="Khá").sum(),(df_TO["Ketqua"]=="Khá").sum(),(df_VL["Ketqua"]=="Khá").sum()]
Trungbinh = [(df_CN["Ketqua"]=="Trung bình").sum(),(df_TO["Ketqua"]=="Trung bình").sum(),(df_VL["Ketqua"]=="Trung bình").sum()]
Yeu = [(df_CN["Ketqua"]=="Yếu").sum(),(df_TO["Ketqua"]=="Yếu").sum(),(df_VL["Ketqua"]=="Yếu").sum()]

df_kq_kh = pd.DataFrame({"Khoa":["CN","TO","VL"],"Giỏi":Gioi,"Khá":Kha,"Trung bình":Trungbinh,"Yếu":Yeu})
df_kq_kh
```

```python
# Bài 19 : Vẽ biểu đồ theo mẫu sau và cho biết các sinh viên có điểm lớn nhất, điểm nhỏ nhất theo mỗi khoa
plt.figure(figsize=(10,5))
sns.swarmplot(x=df_merge["makh"],y=df_merge["diem"])
print("Sinh viên có điểm lớn nhất của khoa CN, khoa TO, khoa VL lần lượt là :",df_CN["diem"].max(),df_TO["diem"].max(),df_VL["diem"].max())
print("Sinh viên có điểm nhỏ nhất của khoa CN, khoa TO, khoa VL lần lượt là :",df_CN["diem"].min(),df_TO["diem"].min(),df_VL["diem"].min())
```

```python
# Bài 20 : Vẽ biểu đồ theo mẫu sau và lập bảng thống kê đếm số nam, số nữ theo mỗi khoa
plt.figure(figsize=(10,5))
sns.swarmplot(x=df_merge["makh"],y=df_merge["diem"],hue=df_merge["gioitinh"]);
```
